```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_wine
from sklearn.metrics import accuracy_score, classification_report
from ucimlrepo import fetch_ucirepo 
```

```python
wine = fetch_ucirepo(id=109)
df = wine.data.features
df['Target'] = wine.data.targets
df.head()
```

```python
# metadata 
print(wine.metadata) 
  
# variable information
print(wine.variables) 
```

```python
X = df.drop('Target', axis = 1)
y = df['Target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
weak_learner = DecisionTreeClassifier(max_depth = 1)
adaboost = AdaBoostClassifier(base_estimator = weak_learner, n_estimators = 50, learning_rate = 1, random_state = 42)
adaboost.fit(X_train, y_train)
y_pred = adaboost.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of AdaBoost on the Wine dataset: {accuracy:.4f}")

classification_rep = classification_report(y_test, y_pred)
print("Classification Report: ", classification_rep)
```
